 _*Merkezi Limit Teoremi Nedir?*_
+ Bir popülasyondan ne kadar çok örnek alınırsa veriler normal dağılıma o kadar yakınsar.
+ Örneklem çapı n =>30 olan değişkenler normal için dağılımdan geldiği varsayımı altında istatistiksel testler uygulanır.
##### Hipotez Testi
+ İstatistiksel hipotez ana kütlenin durumu ile ilgili ortaya atılan
varsayımdır ve örnek istatistikleri kullanılarak test edilir. 
+ H0 Hipotezi yokluk hipotezi olarak
adlandırılır
+ H1 Hipotezi ise alternatif hipotezdir.
+ H0 Ortalamalar arasında fark yoktur.
+ H1: Ortalamalar arasında fark vardır
+ Hipotez testi kurulduktan sonraki aşama test aşamasıdır.
+ Her test içerisinde bir hata payı barındırır.
+ Hata payı alfa (a) olarak adlandırılır.
##### P value
+ P value hata payı olan a ile karşılaştırılarak hipotez testinin kararı
verilir
+ P value değişkenler arasındaki farklılığın şans eseri meydana
gelmesinin olasılığını verir.
+ Eğer p-value a ise HO hipotezi reddedilir.
### Hipotez Türleri
##### Parametrik Hipotez Testleri
+ Gözlem birimleri bağımsız olmalıdır.
+ Veriler normal dağılım gösteren bir yığından gelmelidir
+ Varyanslar homojen olmalıdır.
+ Örneklem çapı n=>30 olmalıdır.
##### Parametrik Olmayan(Nonparametrik) Hipotez Testleri
+ Normal dağılmayan nicel verilerden oluşuyorsa,
+ Birim sayısı yetersizse (TK30),
+ Veri yapısı heterojense
+  Veri seti sıralı ve yaklaşık aralıklı nitel verilerden oluşuyorsa
kullanılır.
##### Parametrik Olmayan (Nonparametrik) Hipotez Testleri Avantajları
+ Dağılım bilgisine gerek yoktur.
+ Gözlem sayısının çok olması gerekmez.
+ Sınıflama ve sıralama verilerinde kullanılabilir.
#####  Dezavantajları
+ Parametrik testlere nazaran güçleri daha düşüktür.
+ Aynı örneklem için farklı parametrik olmayan yöntemler farklı sonuçlar verebilir.

negatif olarak yorumlanabilir.

+ Kovaryans analizi ile yalnızca ilişkinin yönü belirlenebilir.
_*Pearson Korelasyon Katsayısı*_

_*Spearman Korelasyon Katsayısı*_

#####  Temel Bileşenler Analizi vs Faktör Analizi
+ PCA verileri daha az sayıda parçaya indirmek için, faktör analizi ise verilerin altında yatan yapıları anlamak için kullanılır.

##### Diskriminant Analizi
+ Diskriminant analizi gruplar arasındaki ayrıma en fazla etkisi olan ayırıcı özelliğe sahip değişkeni bularak bir sonraki gözlem birimini bir gruba atamaktır.
_*Diskriminant Analizinin Amaçları;*_
+ Gruplar arasındaki farklılıkları analiz etmek,
+ Birbirine benzeyen gözlem birimlerini belirli gruplara ayırmak,
+ Ayrıştırılan gözlem birimlerinin doğru şekilde ayrılıp ayrılmadığını kontrol etmektir.
+ Örnek: Coğrafi bölgeler arasındaki ekonomik farklılıklar, seçim sonuçlarının tahmin edilmesi, bir hastalık üzerinde farklı tedavi yöntemlerinin etkiler vb.
_*Diskriminant Analizinin Varsayımları;*_
+ Veri seti çok değişkenli normal dağılım göstermelidir,
+ Değişkenlerin varyans ve kovaryansları homojen olmalıdır,
+ Değişkenlerin ortalama ve varyansları arasında ilişki/korelasyon bulunmamalıdır,
+ Değişkenler arasında çoklu bağlantı bulunmamalıdır,
+ Birbirinden farklı en az iki grup olmalı ve her grupta en az iki gözlem birimi bulunmalıdır.
##### Diskriminant Analizi
+ Grupları ayırma yöntemi olarak iki farklı yöntem vardır;
+ Doğrusal Yöntem: Gruplar arasında doğrusal çizgiler oluşturarak ayrışım yapar.
+ Kuadratik Yöntem: Gruplar arasındaki ayrışımın doğrusal çizgilerle yapılamadığı durumlarda kullanılan formdur.
